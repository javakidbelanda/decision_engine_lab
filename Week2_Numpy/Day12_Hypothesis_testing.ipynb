```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)

population_mean = 30
population_std = 5

samples = np.random.normal(population_mean, population_std, 1000)
```

```python
sample = np.random.choice(samples, size=40)
sample_mean = np.mean(sample)

print(sample_mean)
```

```python
sample_means = []

for i in range(1000):
    sample = np.random.normal(population_mean, population_std, 40)
    sample_means.append(np.mean(sample))
```

```python
plt.hist(sample_means, bins=40)
plt.axvline(34, color='red')
plt.title("Sampling Distribution Under Null Hypothesis")
plt.show()
print(np.mean(sample_means))
```
